```python
# Create a plot
plt.figure(figsize=(10, 6))

# Plot mass_vector_merged
plt.plot(time_vector_1, mass_vector_1, label='Before Cube', color='blue', marker='o')

# Plot mass_vector_2
plt.plot(time_vector_2, mass_vector_2, label='After Cube', color='green', marker='x')

# Adding titles and labels
plt.title('Mass vs Time')
plt.xlabel('Time')
plt.ylabel('Mass')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
if time_vector_merged.size != mass_vector_1.size:
    flag = True
else:
    flag = False

# Perform linear regression using scipy for mass_vector_1
slope_1, intercept_1, r_value_1, p_value_1, std_err_1 = linregress(time_vector_1, mass_vector_1)

# Generate predictions using the slope and intercept
mass_pred_1 = slope_1 * time_vector_1 + intercept_1

# Plot the regression line
plt.plot(time_vector_1, mass_pred_1, color='blue', label='Before Cube Regression Line')

# Print the regression statistics
print(f"Before Cube: Slope = {slope_1:.4f}, Intercept = {intercept_1:.4f}, R-squared = {r_value_1**2:.4f}")

# Perform linear regression for mass_vector_2
if flag:
    slope_2, intercept_2, r_value_2, p_value_2, std_err_2 = linregress(time_vector_2, mass_vector_2)
    mass_pred_2 = slope_2 * time_vector_2 + intercept_2
    plt.plot(time_vector_2, mass_pred_2, color='green', label='After Cube Regression Line')
    print(f"After Cube: Slope = {slope_2:.4f}, Intercept = {intercept_2:.4f}, R-squared = {r_value_2**2:.4f}")
```

```python
# Create a plot
plt.figure(figsize=(10, 6))

# Plot mass_vector_1
plt.plot(time_vector_1, mass_vector_1, label='Mass 1', color='blue', marker='o')

# Plot the regression line
plt.plot(time_vector_1, mass_pred_1, color='red')

if flag:
    plt.plot(time_vector_2, mass_vector_2, label='Mass 2', color='green', marker='o')
    plt.plot(time_vector_2, mass_pred_2, color='red')

# Adding titles and labels
plt.title('Mass vs Time')
plt.xlabel('Time')
plt.ylabel('Mass')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
